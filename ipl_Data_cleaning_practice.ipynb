SEASON COLUMN DATATYPE IS IN OBJECT FORM WHEN I TRY TO CONVERT IT TO A NUMERIC TYPE DATA TYPE IT IS CONVERTED AS FLOAT. THIS IS HAPPENING BECAUSE THIS COLUMN CONTAIS 2008 AND 2028/09 TYPE DATA 

FOR OVERCOME THIS PROBLEM WE HAVE TO MAKE TO COLUMNS LIKE SEASON START COLUMN AND SEASON END COLUMN

CREATING SEASON START COLUMN

```python
df['Season Start'] = df['Season'].apply(lambda x: x.split('/')[0] if '/' in x else x)
```

CREATING SEASON END COLUMN

```python
df['Season End'] = df['Season'].apply(lambda x: x.split('/')[1] if '/' in x else x)
```

FILLING 08 TYPE OF DATA AS 2008 IN SEASON END COLUMN

```python
# Fix Season End to full year if only 2 digits (like '08' -> '2008')
df['Season End'] = df['Season End'].apply(
    lambda x: '20' + x if len(x) == 2 else x
)
```

```python
df.head(2)
```

POP AND RE-INSERT 'SEASON START' AT THE SECOND COLUMN (INDEX 1)

```python
col = df.pop('Season Start')
df.insert(1, 'Season Start', col)
```

POP AND RE-INSERT 'SEASON END' AT THE SECOND COLUMN (INDEX 2)

```python
col = df.pop('Season End')
df.insert(2, 'Season End', col)
```

```python
df = df.drop('Season', axis=1)
```

```python
df.columns
```

```python
df.dtypes
```

CONVERTING THE DATATYPES OF THE SEASON START AND SEASON END COLUMNS

```python
df['Season Start'] = df['Season Start'].astype(int)
```

```python
df['Season End'] = df['Season End'].astype(int)
```
